## Question 2: Building Text Processing Functions

### Question 2.1: Building a Data Cleaning Function

**Description:** The function removes digits and keeps only the characters "A-Za-z(),!?\'\`".


```python
# clean text
def clean_text(text):
    #### YOUR CODE HERE ####
    pattern = r"[^A-Za-z(),!?'`] "
    return re.sub(pattern, "", text)
    #### END YOUR CODE #####
```

### Question 2.2: Function to Convert Text to Lowercase

```python
# make all text lowercase
#### YOUR CODE HERE ####
def text_lowercase(text):
    return text.lower()
#### END YOUR CODE #####
```

### Question 2.3: Building a Function to Remove Punctuation

```python
# remove punctuation
def remove_punctuation(text):
    #### YOUR CODE HERE ####
    pattern = f"[{re.escape(string.punctuation)}]"
    return re.sub(pattern, "", text)
    #### END YOUR CODE #####
```

### Question 2.4: Tokenization

```python
# tokenize
nltk.download('punkt_tab')
def tokenize(text):
    #### YOUR CODE HERE ####
    tokens = word_tokenize(text)
    return tokens
    #### END YOUR CODE #####
```

### Question 2.5: Removing Stopwords







```python
# remove stopwords
stop_words = set(stopwords.words('english'))
#### YOUR CODE HERE ####
def remove_stopwords(text):
    filtered_tokens = [w for w in text if w.lower() not in stop_words]
    return filtered_tokens
#### END YOUR CODE #####
```

### Question 2.6: Building a Lemmatization Function







```python

from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag

nltk.download('punkt')
nltk.download("averaged_perceptron_tagger_eng")
nltk.download('wordnet')
nltk.download('omw-1.4')

# lemmatize

lemmatizer = WordNetLemmatizer()
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # mặc định là danh từ

def lemmatize(tokens):
    # POS tagging cho từng từ
    pos_tags = pos_tag(tokens)
    # Lemmatize theo POS tag
    lemmatized_words = [
        lemmatizer.lemmatize(word, get_wordnet_pos(tag))
        for word, tag in pos_tags
    ]
    return lemmatized_words

print(lemmatize(["This","is","the","best","film"]))
```

### Question 2.7: Building a Preprocessing Function
**Hint:** This function will call the functions written above.

```python
def preprocessing(text):
    text_cleaned = clean_text(text)
    text_lowercased = text_lowercase(text_cleaned)
    text_removing_punctuation = remove_punctuation(text_lowercased)
    tokens = tokenize(text_removing_punctuation)
    tokens_removing_stopwords = remove_stopwords(tokens)
    lemmatization = lemmatize(tokens_removing_stopwords)
    return lemmatization

#### END YOUR CODE #####
```
